```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
inputG = "valores1.txt"
'''
5 5
1 2 10
1 3 15
2 4 15
3 4 5
4 5 25
'''
```

```python
def readFile(nameFile):
    file = open(inputG,'r'); 
    f = file.readlines(); 
    file.close();
    return f
```

```python
def getG(fileG, fileW = None):
    f = readFile(fileG)
    numV, numEdg = [int(i) for i in f[0].split()]
    edges = [tuple([int(x) for x in edge.split()]) for edge in f[1:]]
    return numV, edges
```

```python
numV, edges = getG(inputG)
edges;
lstN = [ str(i) for i in range(1, numV+1)]
```

```python
G=nx.DiGraph(directed=True)

G.add_nodes_from(lstN)

for edge in edges:
    G.add_edge(str(edge[0]), str(edge[1]), weight=edge[2])
    
pos = nx.spring_layout(G)
weights = nx.get_edge_attributes(G, "weight")
nx.draw_networkx(G, pos, with_labels=True, font_size = 20, font_color = 'white', node_size= 550, node_color='green', width= 3, arrowsize= 30) #VERTICES E ARESTAS
nx.draw_networkx_edge_labels(G, pos, font_size = 20, edge_labels=weights) #PESO DAS ARESTAS
#nx.draw_networkx(G, arrows=True, **options)
plt.rcParams["figure.figsize"] = (15,8)
plt.show()
```
